# Travel Cost Estimation Ai Assistant
### Client Requirements (MAKE MY TRIP)
* our customers enters to and from location while booking a trip
* currently we donot have a system to estimate the travel cost
* can you create an Ai system which takes the {departure} and {Arrival} locations.
* also it takes the {mode of transport} and generate an estimate cost

## Step 1: Import Necessary Libraries

```python
# Step 1: Import libraries
from langchain.prompts import ChatPromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.schema.output_parser import StrOutputParser
```

## Step 2: Design the Template
A template defines how your chatbot will behave — like giving it a personality or role.

```python
chat_template = ChatPromptTemplate(messages=[("system","""Act as a Travel cost Estimation AI Assistant.
                                        you are an AI who takes user input and find the estimates cost of travel.
                                        Make sure not to estimate the cost of staying at the location.
                                        user will enter 3 inputs.
                                                    Inputs 1 : Source Location
                                                    Inputs 2 : Destination Location
                                                    Inputs 3 : Made of Transport
                                                    
                                                    Response Format : Output the cost estimation only.
                                                    do not give any suggestion"""),
                                        ("human","I want to travel to {destination} from {source} via {mode}."),])
chat_template
```

## Step 3: Initialize the Model (Gemini 1.5 / 302-0)

We’ll initialize the Gemini chat model (or you can use OpenAI GPT models too).

```python
#set the api key and initialize a chatmodel
chat_model = ChatGoogleGenerativeAI(api_key="Your Api Key",
                                   model="gemini-2.0-flash-exp",
                                   temperature=1)
```

## Step 4: Design an Output Parser
An Output Parser controls how model responses are returned.

We’ll use a simple string parser here.

```python
output_parser = StrOutputParser()
```

## Step 5: Chaining — Plug It All Together

Combine the prompt + model + parser into one chain. 

```python
# chaining 
chain = chat_template | chat_model | output_parser
```

## Step 6: Test the Chatbot

Now you can chat with your bot directly from the terminal.

```python
source = input("Enter Source: ")
destination = input("Enter Destination: ")
mode = input("Enter Mode of Travel (car/train/bike): ")

user_input = {
    "source": source,
    "destination": destination,
    "mode": mode
}

response = chain.invoke(user_input)
print("\nAssistant:", response)
```
